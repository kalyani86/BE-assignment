```python
def get_text(url):
  response = requests.get(url)
  soup = BeautifulSoup(response.content, 'html.parser')
  text_content = soup.find_all('p')
  text = ' '.join([para.get_text() for para in text_content])
  text = text.lower()
  return text
```
